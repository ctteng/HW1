####3. Use Pandas "Read JSON" functionality to turn the 311 data into a Pandas Dataframe and display the first 20 rows


```python
#YOUR CODE HERE
df_311 = pd.read_json(url)
df_311[0:20]
```

####4. How many missing values do we have? 

```python
#YOUR CODE HERE
len(df_311[df_311['case_id'].isnull()==True])
```

####5. Drop the "point" feature from the dataframe - its a dictionary, which is unsupported in a flat database

```python
#YOUR CODE HERE
df_311.drop('point',axis=1,inplace=True)
```

```python
df_311.drop('media_url',axis=1,inplace=True)
```

####5. Turn the 311 data into a database. You can use the pandas read_sql command (example [here)]('http://bit.ly/1EoP81v') or the db package from yHat to do this. 

```python
#YOUR CODE HERE
df_no_nulls = df_311[df_311['case_id'].isnull()==False]
con = sqlite3.connect(':memory:')
df_no_nulls.to_sql('cases',con=con)
```

###### 6. Display the first 5 rows of the DB

```python
#YOUR CODE HERE
pd.read_sql("SELECT * from cases LIMIT 5", con)
```

####7. Show 10 of the sidewalk cleanings in the Haight Ashbury neighborhood

```python
#YOUR CODE HERE
qry_neigh = """SELECT * 
                From cases 
                Where neighborhood = 'Haight Ashbury'
                And category = 'Street and Sidewalk Cleaning' 
                """
pd.read_sql(qry_neigh, con)
```

####8. Which type of complaint was most common? 

```python
#YOUR CODE HERE
qry_comp = """SELECT count(*) as cat_count, category  
                From cases 
                Group by category 
                Order by cat_count desc 
                Limit 1"""
pd.read_sql(qry_comp, con)
```

####9. Are there any neighborhoods where sidewalk cleaning requests aren't the most common? 

```python
#YOUR CODE HERE
qry = """Select max(cat_count), category, neighborhood From (
            SELECT count(category) as cat_count, category, neighborhood  
            From cases cc
            Where neighborhood != 'None' 
            Group by category, neighborhood 
            Order by neighborhood, cat_count desc
            ) TA
            Where category != 'Street and Sidewalk Cleaning'
            Group By  Neighborhood
            Order by Neighborhood, Category
        """

#print qry
pd.read_sql(qry,con)    
```
